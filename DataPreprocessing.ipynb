# Data Preprocessing

```python
## # 1.Importing required liberiers and packages.
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
# View all columns :
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("E:/project final/cleaned_SC_project_df.csv", encoding='latin1')
df.head()
```

```python
## remove dupliacted data and non add value columns
cols_to_remove = [
    'Order_City',
    'Customer_City',
    'Customer_Country',
    'Customer_State',
    'Order_Country',
    'Customer_FullName',
    'Product_Name',
    'Order_State',
    'Category_Name',
    'same_day_shipping'
]
# Remove only columns that exist in the DataFrame
df = df.loc[:, ~df.columns.isin(cols_to_remove)]
```

```python
df.head()
```

```python
## Using OrdinalEncoder for shipping_mode coilmn as there is relationship betwwen the data and onehotencoder for the rest columms

from sklearn.preprocessing import OrdinalEncoder

ord = OrdinalEncoder(categories= [['Second Class', 'First Class', 'Same Day', 'Standard Class'] ])

df['Shipping_Mode'] = ord.fit_transform(df[['Shipping_Mode']])
df['Shipping_Mode']
```

```python
df.head()
```

```python
# rest categorical coulmns using onehotencoder,

cat_cols = ['Type','Delivery_Status' ,'Customer_Segment','Department_Name','Market','Order_Region','Order_Status']
```

```python
cat_cols
```

```python
for col in cat_cols:
    print(col)
    print(df[col].nunique())
    print(df[col].unique(), '\n', '*' * 100)
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(drop= 'first', sparse_output= False)

ohe_arr = ohe.fit_transform(df[cat_cols])
ohe_arr
```

```python
ohe_df = pd.DataFrame(ohe_arr, columns= ohe.get_feature_names_out())
```

```python
df = pd.concat([df, ohe_df], axis= 1).drop(cat_cols, axis= 1)
df.head()
```

## Split Data into Input Features and Target Column

```python
x = df.drop('Late_delivery_risk', axis= 1)
y = df['Late_delivery_risk']
```

## Split Data into Train and Test

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 42)
```
